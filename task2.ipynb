```python
# Tic Tac Toe game in Python

# Define player symbols
HUMAN_PLAYER = 'X'
COMPUTER_PLAYER = 'O'
EMPTY_SPACE = ' '

# Function to print the game board
def print_game_board(board):
    for row in board:
        print(' | '.join(row))
        print('_' * 10)

# Function to check if a player has won
def check_winner(board, player):
    # Check rows and columns
    for i in range(3):
        if all([cell == player for cell in board[i]]) or all([board[j][i] == player for j in range(3)]):
            return True
    # Check diagonals
    if all([board[i][i] == player for i in range(3)]) or all([board[i][2 - i] == player for i in range(3)]):
        return True
    return False

# Function to check if the game board is full
def is_board_full(board):
    return all([cell != EMPTY_SPACE for row in board for cell in row])

# Function to get empty spaces on the board
def get_empty_spaces(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] == EMPTY_SPACE]

# Minimax algorithm to determine the best move for the computer player
def minimax(board, depth, is_maximizing):
    if check_winner(board, HUMAN_PLAYER):
        return -1
    elif check_winner(board, COMPUTER_PLAYER):
        return 1
    elif is_board_full(board):
        return 0

    if is_maximizing:
        best_score = float('-inf')
        for i, j in get_empty_spaces(board):
            board[i][j] = COMPUTER_PLAYER
            score = minimax(board, depth + 1, False)
            board[i][j] = EMPTY_SPACE
            best_score = max(score, best_score)
        return best_score
    else:
        best_score = float('inf')
        for i, j in get_empty_spaces(board):
            board[i][j] = HUMAN_PLAYER
            score = minimax(board, depth + 1, True)
            board[i][j] = EMPTY_SPACE
            best_score = min(score, best_score)
        return best_score

# Function to find the best move for the computer player
def find_best_move(board):
    best_score = float('-inf')
    best_move = (-1, -1)
    for i, j in get_empty_spaces(board):
        board[i][j] = COMPUTER_PLAYER
        score = minimax(board, 0, False)
        board[i][j] = EMPTY_SPACE
        if score > best_score:
            best_score = score
            best_move = (i, j)
    return best_move

# Function to play the game
def play_game():
    # Initialize the game board
    board = [[EMPTY_SPACE for _ in range(3)] for _ in range(3)]

    while True:
        print_game_board(board)

        # Human player's turn
        row = int(input("Enter row (0, 1, or 2): "))
        col = int(input("Enter column (0, 1, or 2): "))

        if board[row][col] == EMPTY_SPACE:
            board[row][col] = HUMAN_PLAYER
        else:
            print("Cell already occupied. Try again.")
            continue

        if check_winner(board, HUMAN_PLAYER):
            print("You win!")
            break
        elif is_board_full(board):
            print("It's a tie!")
            break

        # Computer player's turn
        print("Computer's turn:")
        print('_' * 10)
        best_move = find_best_move(board)
        board[best_move[0]][best_move[1]] = COMPUTER_PLAYER

        if check_winner(board, COMPUTER_PLAYER):
            print_game_board(board)
            print("Computer wins!")
            break
        elif is_board_full(board):
            print_game_board(board)
            print("It's a tie!")
            break

if __name__ == "__main__":
    play_game()
```
